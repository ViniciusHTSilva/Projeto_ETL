```python
df = df[['name','state', 'brewery_type','website_url']]
```

Definindo os Valores NUlos

```python
df['website_url'] = df['website_url'].fillna('nao informado')
df.head()
```

```python
df = df.dropna()
```

```python
df.count()
```

```python
from sqlalchemy import create_engine
import os
from dotenv import  load_dotenv
load_dotenv()


user = os.getenv("POSTGRES_USER")
password = os.getenv("POSTGRES_PASSWORD")
host = os.getenv("postgres_host")
database = os.getenv("POSTGRES_DB")



engine = create_engine(f"postgresql+psycopg2://{user}:{password}@localhost:5432/{database}")



print("USER:", os.getenv("POSTGRES_USER"))
print("PASSWORD:", os.getenv("POSTGRES_PASSWORD"))
print("HOST:", os.getenv("postgres_host"))
print("DB:", os.getenv("POSTGRES_DB"))
```

```python
df.columns = df.columns.str.lower()

df.to_sql(
    name='final',
    con=engine,
    schema='finaldesemana',
    if_exists='append',
    index=False
)

```
